```python
from google.cloud import vision
import io
import os
import numpy as np
import base64
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image

%matplotlib inline
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "SET YOUR OWN CREDENTIALS'S PATH HERE"
```

```python
image_path = "images/lebron_james.jpg"
```

```python
client = vision.ImageAnnotatorClient()
with io.open(image_path, 'rb') as image_file:
    content = image_file.read()

image = vision.types.Image(content = content)

response = client.text_detection(image = image)
texts = response.text_annotations
```

```python
coordX = []
coordY = []
for text in texts:
    print('\n"{}"'.format(text.description))

    vertices = (['({},{})'.format(vertex.x, vertex.y)
                for vertex in text.bounding_poly.vertices])
    tempX = [vertex.x for vertex in text.bounding_poly.vertices]
    tempY = [vertex.y for vertex in text.bounding_poly.vertices]
    
    coordX.append(tempX)
    coordY.append(tempY)

    print('bounds: {}'.format(','.join(vertices)))
```

```python
import cv2

image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("My image")
plt.show()
```

```python
for idx, X in enumerate(coordX):
    # draw the bounding box on the image
    pts = np.array([[X[0],coordY[idx][0]],
                    [X[1],coordY[idx][1]],
                   [X[2],coordY[idx][2]],
                   [X[3],coordY[idx][3]]], np.int32)
    pts = pts.reshape((-1,1,2))
    cv2.polylines(image, [pts], True, (0, 255, 0), 2)

plt.imshow(image)
plt.title("My image")
plt.show()
```
